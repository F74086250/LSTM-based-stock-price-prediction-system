```python
#切分Test集
test = data[data.date>'2021-02-01']
train = data[:len(data)-len(test)]
#只要open high
arr1_train = train["open"].values.reshape(-1,1)
arr1_test = test["open"].values.reshape(-1,1)
arr2_train=train["Trading_Volume"].values.reshape(-1,1)
arr2_test=test["Trading_Volume"].values.reshape(-1,1)
train_set=np.hstack((arr1_train,arr2_train))
test_set=np.hstack((arr1_test,arr2_test))
test_set
```

```python
from sklearn.preprocessing import MinMaxScaler 
sc = MinMaxScaler(feature_range = (0, 1))
#需將資料做reshape的動作，使其shape為(資料長度,1) 
#train_set= train_set.values.reshape(-1,1)
training_set_scaled = sc.fit_transform(train_set)
training_set_scaled
```

```python
X_train = [] 
y_train = []
for i in range(10,len(train_set)):
    X_train.append(training_set_scaled[i-10:i, 0]) 
    y_train.append(training_set_scaled[i, 0]) 
X_train, y_train = np.array(X_train), np.array(y_train) 
X_train = np.reshape(X_train, 
                         (X_train.shape[0], X_train.shape[1], 1))

```

```python
print("第i天之前的股價")
print(X_train[0])
print("第i天股價")
print(y_train[0])
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout,BatchNormalization
```

```python
keras.backend.clear_session()
regressor = Sequential()
regressor.add(LSTM(units = 100, input_shape = (X_train.shape[1],1),return_sequences="TRUE"))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 100,activation='ReLU'))
regressor.add(Dense(units = 1))
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics=[[keras.metrics.MeanSquaredError()]])
```

```python
regressor.summary()
```

```python
history = regressor.fit(X_train, y_train, epochs =100, batch_size = 16)
plt.title('train_loss')
plt.ylabel('loss')
plt.xlabel('Epoch')
plt.plot( history.history["loss"])
```

```python
dataset_total = pd.concat((train['open'], test['open']), axis = 0)
inputs = dataset_total[len(dataset_total) - len(test) - 10:].values
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
X_test = []
for i in range(10, len(inputs)):
    X_test.append(inputs[i-10:i-1, 0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_stock_price = regressor.predict(X_test)
#使用sc的 inverse_transform將股價轉為歸一化前
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
plt.plot(test['open'].values, color = 'black', label = 'Real TSMC Stock Price')
plt.plot(predicted_stock_price, color = 'green', label = 'Predicted TSMC Stock Price')
plt.title('TATA Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
# plt.savefig('lstm_2330.png')
```

```python
scores = regressor.evaluate(X_test)
scores
```
